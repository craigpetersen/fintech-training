<h2>Functions</h2> 

Functions are defined using the <bold>def</bold> keyword. Functions in Python are first class citizens and can be passed into other functions etc.  

<p>Python is a "white-space" language i.e. white-space is important to the interpreter.  The default number of whitespace chars separating nested statements is typically four spaces - one tab stop.  Note the function below is declared with "is_prime" name, it takes one argument called n.  It also returns a boolean (True/ False) and this is meant to indicate the "primality" of the number.  Interestingly, the n-arg can be of any type.  So it could be a string or a list of integers etc.  This is because Python is "duck typed" - not  statically typed so it cannot check prior to runtime what types should be passed.  This is a double-edged sword. </p>

```python
def is_prime(n):
    if n == 1:
        return False
    if n == 2:
        return True
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(10))
# print(is_prime("11"))  # This will throw an error
```

Comprehension lists are fast and easy ways to loop through data.  Typically, they are used within the list format "[...]" so return a list.

```python
data = [(x,y) for x in  range(1,100) for y in range(1,100) if is_prime(y)]
plt.plot(data)
```
